## Mini Hundir la Flota
#### Primeros pasos
Vamos a empezar a programar trozos de código que nos servirán para el HLF:

* Crear un Tablero como una lista de listas de 10x10 que contenga caracteres espacio: " "
* Insertar dos barcos sustituyendo en el Tablero los espacios en blanco por el caracter "B":
    * Uno de 4 casillas en horizontal: casillas [1][0], [1][1], [1][2], [1][3]
    * Uno de 3 casillas en vertical:   casillas [3][3], [4][3], [5][3]
* Escribir una celda de código separada de las anteriores para simular que el usuario está efectuando un disparo: 
    *  Solicite al usuario dos coordenadas `i,j` mediante dos inputs
    *  Compruebe si esas coordenadas están ocupada por una casilla barco (que haya una en "B")
        * Si hay una "B":
            * Cambiar el contenido de esa posición por una "x"
            * Imprimir por pantalla el texto "Tocado en posición `i,j`"
        * Si hay un " ":
            * Cambiar el contenido de esa posición por una "o"
            * Imprimir por pantalla el texto "Agua"
    * Imprimir por pantalla el Tablero
* Para comprobar que funciona bien la celda de disparar ejecutarla varias veces metiendo distintas coordenadas

### Bonus
* En la celda de disparar, si el usuario acierta a tocar un barco, vuelve a disparar
* Resuélvelo con funciones:
    * Una para solicitar una coordenada válida a la que llamarás dos veces, una par i y otra par j
    * Una `disparar` que devuelva `True` si el usuario ha acertado o `False` si no.
    


```python
# Creamos un tablero de 10x10 lleno de espacios " "
tablero = []

for i in range(10):
    fila = []
    for j in range(10):
        fila.append(" ")
    tablero.append(fila)

# Mostramos el tablero para comprobar
for fila in tablero:
    print(fila)

 #Barco horizontal de 4 casillas en:
# [1][0], [1][1], [1][2], [1][3]
tablero[1][0] = "B"
tablero[1][1] = "B"
tablero[1][2] = "B"
tablero[1][3] = "B"

# Barco vertical de 3 casillas en:
# [3][3], [4][3], [5][3]
tablero[3][3] = "B"
tablero[4][3] = "B"
tablero[5][3] = "B"

# Mostramos el tablero para comprobar
for fila in tablero:
    print(fila)

# Pedimos coordenadas al usuario
i = int(input("Introduce la coordenada i (fila, 0-9): "))
j = int(input("Introduce la coordenada j (columna, 0-9): "))

# Comprobamos el contenido de esa casilla
if tablero[i][j] == "B":
    tablero[i][j] = "x"
    print(f"Tocado en posición {i},{j}")
elif tablero[i][j] == " ":
    tablero[i][j] = "o"
    print("Agua")
else:
    # Por ejemplo, ya había una 'x' o una 'o'
    print("Ya has disparado antes en esa posición")

# Imprimimos el tablero después del disparo
for fila in tablero:
    print(fila)
    

def pedir_coordenada(nombre):
    """
    Pide una coordenada (i o j) entre 0 y 9.
    'nombre' es solo para mostrar en el mensaje (por ejemplo, 'fila' o 'columna').
    """
    while True:
        try:
            valor = int(input(f"Introduce la coordenada de {nombre} (0-9): "))
            if 0 <= valor <= 9:
                return valor
            else:
                print("La coordenada debe estar entre 0 y 9.")
        except ValueError:
            print("Debes introducir un número entero.")
```
